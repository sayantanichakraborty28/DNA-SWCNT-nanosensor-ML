```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv('heatmap-inp.csv', header=None, names=['Row', 'Column', 'Value'])

# Pivot the DataFrame to create a matrix
matrix_df = df.pivot(index='Row', columns='Column', values='Value')

# Create the heatmap
sns.set(font_scale=1.2)
plt.figure(figsize=(10, 8))

# Colormap similar to your uploaded figure (purpleâ†’yellow)
cmap = sns.color_palette("inferno_r", as_cmap=True)

heatmap = sns.heatmap(matrix_df, annot=True, cmap=cmap, fmt='d',
                      linewidths=.5, square=True, annot_kws={"size": 28},
                      cbar_kws={'label': '# of molecules'})

# Adjust colorbar tick size
cbar = heatmap.collections[0].colorbar
cbar.ax.tick_params(labelsize=18)

# Labels
plt.ylabel('ML-predicted response', fontsize=28, font='Arial')
plt.xlabel('Experimental response', fontsize=28, font='Arial')

plt.xticks(fontsize=22)
plt.yticks(fontsize=22)

plt.savefig('heatmap-en-newmodels.png', dpi=300, bbox_inches='tight')
plt.show()
```
